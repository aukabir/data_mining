```python
import pdfplumber
import pandas as pd
from PIL import Image
```

```python
pdf = pdfplumber.open('/content/warn_report.pdf')

#Load page_0
p0 = pdf.pages[0]

# Visualize the actual content on page_0
Image.open('/content/page_0.png')
```

```python
table = p0.extract_table()

df = pd.DataFrame(table[1:], columns=table[0])
for column in ["Effective", "Received"]:
    df[column] = df[column].str.replace(" ", "")
```
